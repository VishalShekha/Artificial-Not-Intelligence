Data Science Use Cases:

1. Data availability: historical sales etc
2. Data Characteristics: time series data
3. Data Quality: Data cleaning not required
4. Modeling: Linear regression
5. Challenge: Abnormal sales trend


Time Series Data:
1. Trend: Sales increasing or decreasing
2. Seasonality: Over a period of time the trends tend to repeat
3. Noise: Uncertainity of data

Bias Variance Tradeoff:
1. Bias: Error during Training
2. Variance: Error during Testing

Cross-Validation:
Step-by-step process of model building which ensure that the model fits the data accurately and does not overfit
Thus prevents overfitting and underfitting

Major Types of Cross Validation and implementation:

1. K-Fold Cross Validation:
Divide the data into K-equal parts such that each part represents the broad dataset.
It trains the dataset in such a way that it uses K-1 parts for training and 1 part of testing. Low Bias and variance ensure that mode lis not learning any noise

2. Stratified K-Fold Cross Validation:
In case of binary classification our target variable may be skewed ( 90% +ve, 10% -ve). Stratified K-fold ensures that each of the k-train and test sample have same label ratio

3. Hold -Out Based Validation:
When amount of dataset is very large say 1 mil the computational cost will increase exponentially. Small hold out set is kept out of training and used fro testing. Used for time series-based data
